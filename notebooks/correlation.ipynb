```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/cleaned_telco.csv')
print(df.shape)
df.describe()
```

```python
df['Churn_Binary'] = df['Churn'].map({'Yes': 1, 'No': 0})
df['TechSupport_Binary'] = df['TechSupport'].map({'Yes': 1, 'No': 0})
```

```python
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix, cmap='Greys', annot=True)
plt.show()  
```

```python
df_imp = df[['PaymentMethod', 'InternetService', 'TechSupport', 'Contract', 'tenure', 'MonthlyCharges', 'Churn']]
df_encoded = pd.get_dummies(df_imp, drop_first=True)
corr = df_encoded.corr()
```

```python
plt.figure(figsize=(14,10))
sns.heatmap(corr, annot=True, cmap='coolwarm', annot_kws={"size":8})
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.show()
```
